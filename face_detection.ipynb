```python
import cv2
from matplotlib import pyplot as plt
import dlib
import numpy as np
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX

dirs = "/home/fabio/.pyenv/versions/3.6.1/lib/python3.6/site-packages/cv2/data/"
```

```python
cascPath = dirs + "haarcascade_frontalface_default.xml"
eyePath = dirs + "haarcascade_eye.xml"
smilePath = dirs + "haarcascade_smile.xml"
```

```python
faceCascade = cv2.CascadeClassifier(cascPath)
eyeCascade = cv2.CascadeClassifier(eyePath)
smileCascade = cv2.CascadeClassifier(smilePath)
```

```python
gray = cv2.imread('images/Sergio.jpg', 0)

plt.figure(figsize=(12,8))
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
faces = faceCascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors = 5, flags = cv2.CASCADE_SCALE_IMAGE)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(gray, (x, y), (x+w, y+h), (255, 255, 255), 3)
```
